There is alot of great information here and luckily most of it can be joined using related columns. Let's start with our first request to find the total sales and number of purchases per country.

```python
%%sql

SELECT
    BillingCountry AS 'Billing Country',
    '$' || sum(total) AS 'Total Sales',
    COUNT(total) AS 'Num of Sales'
FROM
    invoices
GROUP BY
    BillingCountry
ORDER BY
    sum(total) DESC;
```

The next one is a bit trickier since I will have to join multiple tables in order to get the top Sales from employees. Also, I'm having to assume RepID and EmployeeID are the same. If confirmed, I could rename one of the columns but for now I will leave it.

```python
%%sql

SELECT
    e.FirstName AS 'First Name',
    e.LastName AS 'Last Name',
    '$' || ROUND(SUM(i.Total), 2) AS 'Total'
FROM
    invoices AS i
INNER JOIN
    customers AS C
ON
    i.CustomerID = c.CustomerID
INNER JOIN
    employees AS e
ON
    c.SupportRepId = e.EmployeeID
GROUP BY
    e.FirstName
ORDER BY
    i.Total DESC;
```

Now let's look for the highest earningsongs in the US. Notice the distiction that we are looking for overall profit of the songs vs physical number of individual purchases. Since each song can range in single purchase value, the highest total sales doesn't necessarily equal highest total download count. 

For this I will need invoices, invoice_items, and tracks. I used inner join since I only want relevant data without null values. It's important to ORDER BY the sum total instead of just total as the data will not be accurate. The only other changes were cleaning up the results by changing column names, adding dollar symbol, and rounding 2 decimal places.

```python
%%sql

SELECT
    t.Name AS 'Song',
    SUM(ii.Quantity) AS 'Quantity',
    '$' || t.UnitPrice AS 'Unit Price',
    '$' || ROUND(SUM(ii.Quantity) * t.UnitPrice, 2) AS 'Sales'
FROM
    invoices AS i
INNER JOIN
    invoice_items AS ii
ON
    i.InvoiceId = ii.InvoiceId
INNER JOIN
    tracks AS t
ON
    ii.TrackId = t.TrackId
GROUP BY
    t.Name
ORDER BY
    SUM(ii.Quantity) DESC
LIMIT
    20;
```

Moving on we will look at sales by state.

```python
%%sql

SELECT
    BillingState,
    '$' || ROUND(SUM(Total), 2) AS Total
FROM 
    invoices
WHERE
    BillingCountry = 'USA'
GROUP BY
    BillingState
ORDER BY
    SUM(Total) DESC;
```

Finally we are going to look up the top purchasers

```python
%%sql

SELECT
    CustomerId,
    '$' || ROUND(SUM(Total), 2) AS Total
FROM 
    invoices
WHERE
    BillingCountry = 'USA'
GROUP BY
    CustomerId
ORDER BY
    SUM(Total) DESC
LIMIT
    10;
```
